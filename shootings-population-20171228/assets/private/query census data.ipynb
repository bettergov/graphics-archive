```python
import requests
import os
from pprint import PrettyPrinter
import pandas as pd

places_df = pd.read_csv('cook-county-places.csv')
places_string = ",".join([str(x) for x in places_df['place'].values])

payload = {
    'NAME': 'Name',
    'P0030001': 'Total population',
    'P0030002': 'White alone',
    'P0030003': 'Black or African American alone',
    'P0030004': 'American Indian and Alaska Native alone',
    'P0030005': 'Asian alone',
    'P0030006': 'Native Hawaiian and Other Pacific Islander alone',
    'P0030007': 'Some Other Race alone',
    'P0030008': 'Two or More Races'
}
payload_string = ",".join(list(payload.keys()))

pp = PrettyPrinter(indent=4)

params = {'get': payload_string,
         'for': 'place:%s' % places_string,
#          'in': 'county:031 state:17',
         'in': 'state:17',
         'key': os.environ['US_CENSUS_API_KEY']}

r = requests.get('https://api.census.gov/data/2010/sf1', params=params)
```

```python
census_df = pd.DataFrame(eval(r.text)[1:], columns=eval(r.text)[0])
census_df = census_df.rename(columns=payload)
display(census_df)
```

```python
round(census_df['White alone'].astype('int64')/census_df['Total population'].astype('int64'), 2)
```

```python
for col in census_df[census_df.columns[2:9]]:
    census_df['Percent %s' % col] = round(census_df[col].astype('int64')/census_df['Total population'].astype('int64'), 2)
```

```python
import requests
import os
from pprint import PrettyPrinter
import pandas as pd

places_df = pd.read_csv('cook-county-places.csv')
places_string = ",".join([str(x).zfill(5) for x in places_df['place'].values])
# places_string = '*'

# Variables for ACS 2016 Profile: https://api.census.gov/data/2016/acs/acs5/variables.html
payload = {
    'NAME': 'Name',
    'DP05_0028E': 'Total population',
#     'DP05_0032PE': 'White only',
#     'DP05_0033PE': 'Black or African American only',
#     'DP05_0034PE': 'American Indian and Alaska Native only',
#     'DP05_0039PE': 'Asian only',
#     'DP05_0047PE': 'Native Hawaiian and Other Pacific Islander only',
#     'DP05_0052PE': 'Some other race only',
#     'DP05_0053PE': 'Two or more races only',
    'DP03_0062E': 'Median household income',
    'DP03_0119PE': 'Percent in poverty'
}
payload_string = ",".join(list(payload.keys()))

pp = PrettyPrinter(indent=4)

params = {'get': payload_string,
         'for': 'place:%s' % places_string,
#          'in': 'county:031 state:17',
         'in': 'state:17',
         'key': os.environ['US_CENSUS_API_KEY']}

r = requests.get('https://api.census.gov/data/2016/acs/acs5/profile', params=params)
```

```python
census_df = pd.DataFrame(eval(r.text)[1:], columns=eval(r.text)[0])
census_df = census_df.rename(columns=payload)
# census_df['Median household income'] = census_df['Median household income'].astype('int64')
# display(census_df.sort_values(by=['Median household income'], ascending=False))
```

```python
for col in census_df:
    try:
        census_df[col] = pd.to_numeric(census_df[col])
    except ValueError:
        pass
```

```python
census_df['Type'] = census_df['Name'].map(lambda x: x.split(' ')[-2][:-1])
census_df['Name'] = census_df['Name'].map(lambda x: ' '.join(x.split(' ')[:-2]))
census_df
```

```python
census_df.to_csv('cook-county-poverty.csv', index=False)
```
