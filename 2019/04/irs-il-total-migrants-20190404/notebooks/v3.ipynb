```python
import os
import pandas as pd
```

```python
scope = sorted(list(filter(lambda x: '_in_' in x, os.listdir('working'))))
scope
```

```python
dfs = []

for filename in scope:
    state, direction, date_from, date_to = filename.split('.')[0].split('_')
    df = pd.read_excel('working/' + filename)
    df['direction'] = direction
    df['date_from'] = date_from
    df['date_to'] = date_to
    df = df[df['Number of returns'].notnull()]
    dfs.append(df)
    
pd.concat(dfs)
```
