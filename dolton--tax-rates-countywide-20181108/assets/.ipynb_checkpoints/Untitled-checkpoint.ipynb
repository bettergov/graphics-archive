```python
full_text_list = list(filter(None, full_text.splitlines()))

def get_line_number(town_name):
    try:
        return list(
            filter(
                lambda x: town_name == x[1].strip(),
                enumerate(full_text_list)
            )
        )[0][0]
    except:
        return None

def get_2017_rate(town_name):
    town_line = list(
        filter(
            lambda x: town_name == x[1].strip(),
            enumerate(full_text_list)
        )
    )[0]
    
    return full_text_list[town_line[0] + 1]

# print(towns[62])
# get_2017_rate(towns[62])

# [(idx, line) for idx, line in full_text.splitlines() if towns[0] in line]
```

```python
import pandas as pd

df = pd.DataFrame(towns, columns=['name'])
df['line_number'] = df.name.apply(get_line_number).fillna(0).astype(int)
df['rate_2017'] = df.line_number.apply(lambda x: full_text_list[x + 1])
df
```

```python
df = df.append({'name': 'CITY OF CHICAGO', 'line_number': -1, 'rate_2017': 7.266}, ignore_index=True)
df.name = df.name.str.replace('VILLAGE OF ', '').str.replace('CITY OF ', '').str.replace('TOWN OF ', '')

df.to_csv('cook-county-tax-rates.csv', index=False)
```
